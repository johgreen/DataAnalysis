```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
def multi_plots(ticker_ , times_):
    '''
        Creates a series of plots of the opening prices of the ticker_. 
        times_ should be a list of dates (as strings) in the form yyyy-mm-dd.
    '''
    tick = yf.Ticker(ticker_)
    
    fig , ax = plt.subplots(len(times_)-1,1,figsize = (10,10))

    for i,axes in zip(times_, ax):
        if times_.index(i) == 0:
            dftt_ = tick.history(start = times_[times_.index(i)] , end = times_[times_.index(i) + 1])
            axes.plot(dftt_.index , dftt_.Open)
            axes.set_title("{} {} - {}".format(ticker_, times_[times_.index(i)] , times_[times_.index(i) + 1]))
            axes.set_ylabel("Price [USD]")
        elif times_.index(i) < len(times_) - 1:
            dftt_ = tick.history(start = times_[times_.index(i)] , end = times_[times_.index(i) + 1])
            axes.plot(dftt_.index, dftt_.Open)
            axes.set_title("{} {} - {}".format(ticker_, times_[times_.index(i)] , times_[times_.index(i) + 1]))
            axes.set_ylabel("Price [USD]")
    fig.tight_layout()
    
```

```python
def dates(starting_date, time_interval, num_intervals):
    '''
        input date in the form "yyyy-mm-dd" and provide an interval in years. 
        this will return the dates, seperated by the interval in a list.
        time_interval must be in years.
    '''
    final_list = []
    for i in range(0,num_intervals):
        if i == 0:
            t1 = str(starting_date)
            final_list.append(t1)
        else:
            t1 = str(int(t1[0:4]) + int(time_interval)) + "-" + t1[5:]
            final_list.append(t1)
    return final_list
```

```python
date_list = dates("2010-01-01" , 2 , 5)
multi_plots("SPY", date_list)
```
